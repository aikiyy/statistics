```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import comb
```

```python
n = 100 #試行回数
p = 0.9 #事象が起こる確率

x_list = pd.Series([comb(float(n), x)*p**x*(1-p)**(float(n)-x) for x in range(0, n+1)])
```

```python
plt.bar(x_list.index,x_list)
plt.xlim(0,100)
plt.xticks(np.arange(0,n+1,10))
plt.xlabel('ある事象が起こる回数')
plt.ylabel('その確率')
plt.show()
```
