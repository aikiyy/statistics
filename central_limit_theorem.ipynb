```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generage_exponential_list(lam, n):
    return np.random.exponential(1/lam, n)
```

# 指数分布を描画する

```python
lam = 0.01
n = 1000

sample = generage_exponential_list(lam, n)
print('標本平均は{}'.format(sample.mean()))

sns.distplot(sample, bins=30, kde=False, hist_kws={"rwidth":0.9,'edgecolor':'black'})
plt.title('サイズ={}, λ={}の指数分布'.format(n, lam))
plt.show()
```

# リサンプリング回数を変えて標本平均分布を描画する

```python
for repean_n in [100, 1000, 10000, 100000]:
    sample_mean = np.array([generage_exponential_list(lam, n).mean() for i in range(repean_n)])
    print('標本平均の平均は{}'.format(sample_mean.mean()))

    sns.distplot(sample_mean, bins=30, kde=False, hist_kws={"rwidth":0.9,'edgecolor':'black'})
    plt.title('λ={}の指数分布に従うサイズ={}の標本平均を{}回リサンプリングした分布'.format(lam, n, repean_n))
    plt.show()
```
