**3分布の関係**
- ガンマ分布: 単位時間あたりに平均λ回発生するランダムな事象が、x時間にα回発生する確率
- ポワソン分布: 単位時間あたりの生起確率（回数に着目）
- 指数分布: 事象の生起間隔の確率（時間に着目）
    
**参考リンク**
- [指数分布とポアソン分布のいけない関係](https://www.slideshare.net/teramonagi/ss-11296227)

```python
import numpy as np
from scipy import stats
```

```python
def generate_gamma_list(a, b, n):
    return np.random.gamma(a, scale=1/b, size=n)
```

# ガンマ分布の描画
gamma関数への引数でかかせないのは以下の2つ
- shape : 何回起こるか、すなわち↑でいうλ
- scale  : どのくらいの期間か、すなわち↑でいうx

[scipy.stats.gamma — SciPy v1.2.1 Reference Guide](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html)

```python
a = 20 # shape
b = 2 # 逆数がscale
n = 1000 # size

# scipyによるpdf
x = np.linspace(0, 50, 100)
y = stats.gamma.pdf(x, a, scale=1/b)
plt.plot(x, y)

# randomで生成したnormalized density
plt.hist(generate_gamma_list(a, b, n), bins=30, normed=1)

plt.show()
```

# ガンマ分布とポワソン分布
- ガンマ分布のscaleを1に固定したものがポワソン分布

[scipy.stats.poisson — SciPy v1.2.1 Reference Guide](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html)

```python
a = 20 # shape
b = 1 # 逆数がscale

# scipyによるgamma pdf
x = np.linspace(0, 50, 100)
y = stats.gamma.pdf(x, a, scale=1/b)
plt.plot(x, y)

# scipyによるpoisson pdf
mu = a / b
x = np.arange(stats.poisson.ppf(0.01, mu), stats.poisson.ppf(0.99, mu))
y = stats.poisson.pmf(x, mu)
plt.plot(x, y)

plt.show()
```

# ガンマ分布と指数分布
- ガンマ分布のshapeを1に固定したものが指数分布

```python
a = 1 # shape
b = 2 # 逆数がscale
n = 1000 # size

# scipyによるgamma pdf
x = np.linspace(0, 10, 100)
y = stats.gamma.pdf(x, a, scale=1/b)
plt.plot(x, y)

# scipyによるexponential pdf
y = stats.expon.pdf(x, scale=1/b)
plt.plot(x, y)

plt.show()
```
